### Inter-quarter range

```python
# Function to calculate Q1, Q3, IQR, Lower Bound, and Upper Bound
def calculate_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return pd.Series({
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': Q1 - 1.5 * IQR,
        'Upper Bound': Q3 + 1.5 * IQR
    })

# Apply the function to numeric columns only
iqr_report = df.select_dtypes(include=['float64', 'int64']).apply(calculate_iqr)

# Transpose to get columns as keys and IQR details as values
iqr_report = iqr_report.T

# Display the resulting IQR report
iqr_report
```

### Calculating percentile
